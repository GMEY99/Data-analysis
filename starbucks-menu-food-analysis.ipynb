```python
import numpy as np   
import matplotlib.pyplot as plt  
import pandas as pd  
import seaborn as sns
```

```python
import pandas as pd
import chardet

# Detect the encoding
with open(r'C:\Users\gmeyg\Downloads\Compressed\starbucs\starbucks-menu-nutrition-food.csv', 'rb') as f:
    result = chardet.detect(f.read())

# Use the detected encoding
df = pd.read_csv(r'C:\Users\gmeyg\Downloads\Compressed\starbucs\starbucks-menu-nutrition-food.csv', encoding=result['encoding'])
```

```python
df
```

```python
df=df.rename(columns={'Unnamed: 0':'menu'})
df=df.rename(columns={' Calories':'Calories'})
df=df.rename(columns={' Fat (g)':'Fat(g)'})
df=df.rename(columns={' Carb. (g)':'Carb(g)'})
df=df.rename(columns={' Fiber (g)':'Fiber(g)'})
df=df.rename(columns={' Protein (g)':'Protein(g)'})
```
